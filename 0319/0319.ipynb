## HW7

```python
def calculator(a,b,c):
    if(c=="+"):
        return(int(a)+int(b))
    if(c=="-"):
        return(int(a)-int(b))
    if(c=="*"):
        return(int(a)*int(b))
    if(c=="/"):
        return(int(a)/int(b))
x=input("請輸入兩個數字加減乘除")
if x.find("+")!=-1:
    a=x.split("+")
    a.append("+")
    print(calculator(a[0],a[1],a[2]))
elif x.find("-")!=-1:
    a=x.split("-")
    a.append("-")
    print(calculator(a[0],a[1],a[2]))
elif x.find("*")!=-1:
    a=x.split("*")
    a.append("*")
    print(calculator(a[0],a[1],a[2]))
else:x.find("/")!=-1
a=x.split("/")
a.append("/")
print(calculator(a[0],a[1],a[2]))
```
